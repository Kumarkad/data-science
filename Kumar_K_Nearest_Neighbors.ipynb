```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sbn
```

```python
df=pd.read_csv("https://github.com/Kumarkad/Dataset/raw/main/adult.csv")
```

# Part 1 : Understanding / Cleaning the Data

```python
df.head()
```

```python
df.info()
```

```python
df=df.replace('?',np.NaN)
df.isnull().sum()
```

```python
df=df.dropna(axis=0)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df['income'].value_counts(normalize=True)
```

```python
df['income'] = (df['income']==("<=50K")).astype(int)
```

```python
df['income'].value_counts(normalize=True)
```

```python
df_str=df.select_dtypes(include='object')
df_int=df.select_dtypes(exclude='object')
```

```python
df_str.describe(include='all')
```

```python
df_str.head()
```

```python
df_int.head()
```

```python
df_str=pd.get_dummies(df_str)
```

```python
df_str.head()
```

# Part 2 : EDA

```python
sbn.pairplot(df_int)
```

```python
plt.figure(figsize=(16,6))
mask=np.triu(np.ones_like(df_int.corr(),dtype=bool))
heatmap=sbn.heatmap(df_int.corr(),mask=mask,vmin=1,vmax=1,annot=True,cmap='BrBG')
heatmap.set_title('Triangle correlation Heatmap',fontdict={'fontsize':18},pad=16);
```

# Part 3 : Standard-Scaling / Max-Min scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
y=df_int['income']
X=df_int.drop(columns='income')
column_names=X.columns.values
```

```python
ss=StandardScaler()
X=ss.fit_transform(X)
```

```python
X=pd.DataFrame(X)
X.columns=column_names
X.head()
```

```python
X=pd.concat([X,df_str],axis=1)
```

# Part 4 : Training and Modeling

```python
from sklearn.model_selection import train_test_split 
x_train,x_test, y_train, y_test =train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier().fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

# Part 5 : Choosing an appropriate K value

```python
error_rate=[]
for i in range(1,50):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i=knn.predict(x_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(1,50),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('Error Rate vs K value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

# Part 6 : Retain with new K value

```python
knn=KNeighborsClassifier(n_neighbors=29)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
